```python
%matplotlib inline
#Dependencies and Setup
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D
import pandas as pd
import numpy as np
```

```python
nba_data = '../Potential Data Sets/NBA Free Throw Data.csv'
nba_data_df = pd.read_csv(nba_data)
```

```python
nba_data_df.head(30)
```

```python
nba_data_df.rename(columns={'Season': 'League Year'})
```

```python
teams_data = nba_data_df.groupby('Team')
```

```python
Team_summation_data_df = pd.DataFrame({'FT Made TOT': teams_data['FT Made'].sum(), 'FT ATT TOT': teams_data['FT Att'].sum(),\
                                        'FT Made% TOT': round(teams_data['FT Made'].sum()/teams_data['FT Att'].sum()*100,2),\
                                       'AVG Win Rnk': round(teams_data['Win Rnk'].mean(),0), 'FT Made Rnk TOT':\
                                       round(teams_data['AVG FT Made Rnk'].mean(),0), 'FT ATT Rnk TOT': round(teams_data['FT ATT Rnk'].mean(),0),\
                                        'PTS TOT': teams_data['Total Points'].sum()})
```

```python
Team_summation_data_df
```

```python
x = Team_summation_data_df['FT Made Rnk TOT']
y = Team_summation_data_df['FT ATT Rnk TOT']

plt.scatter(x,y, color='darkorange', marker='o', )

z = np.polyfit(x,y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"k--",  markeredgecolor='black', fillstyle='bottom', markersize=15, markerfacecoloralt='black' )

plt.title('FT Made Rnk TOT vs. FT ATT Rnk TOT')
plt.xlabel('FT Made Rnk TOT') 
plt.ylabel('FT ATT Rnk TOT')

plt.grid()
plt.savefig('FT Made Rnk TOT vs. FT ATT Rnk TOT.png')
plt.show()
```

```python
x = Team_summation_data_df['FT Made TOT']
y = Team_summation_data_df['Win Rnk TOT']
plt.scatter(x,y,  color='darkorange', marker='o')

z = np.polyfit(x,y, 1)
p = np.poly1d(z)
plt.plot(x,p(x), "k--")
plt.title('FT Made TOT VS. Win Rnk TOT')
plt.xlabel('FT Made TOT')
plt.ylabel('Win Rnk TOT')
plt.grid()
plt.savefig('FT Made TOT VS. Win Rnk TOT.png')
plt.show()
```
